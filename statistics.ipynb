````python
#Question 3:  Write a Python class representing a discrete random variable with methods to calculate its expected value and variance?

#ans:
'''expected_value method: Computes the expected value of the random variable by summing 
i . i for all outcomes

variance method :i^2 . i'''

class DiscreatRandomVariable:
    def __init__(self,outcomes,probabilities):
        #Ensure that lenth of outmes and probabilities must match

        if len(outcomes)!=len(probabilities):
            raise ValueError("the number of outcomes and probabilites must match")
        
        #Ensure that probabilities sum to 
        if not abs(sum(probabilities)-1.0)<1e-6:
            raise ValueError("The probalities must sum to 1.")
        
        self.outcomes=outcomes
        self.probabilities=probabilities

    def expected_value(self):
        """ Calculate the expexcted value E[X]"""
        return sum(x*p for x,p in zip(self.outcomes,self.probabilities))
    
    def variance(self):
        """calculating the expecteed value Var[X]"""
        #first calculate E[X]
        expected_value=self.expected_value()

        #Now calculate E[X^2]
        expected_value_of_square=sum((x**2)*p for x,p in zip(self.outcomes,self.probabilities))

        #Variance is E[X^2]-E[X]^2
        return expected_value_of_square - expected_value
         
# Example Usage:
outcomes = [1, 2, 3, 4, 5]
probabilities = [0.1, 0.2, 0.3, 0.2, 0.2]  # Probabilities must sum to 1

# Create a discrete random variable object
drv = DiscreatRandomVariable(outcomes, probabilities)

# Calculate the expected value and variance
expected_value = drv.expected_value()
variance = drv.variance()

print(f"Expected Value: {expected_value}")
print(f"Variance: {variance}")

````

````python
#Question 4: Implement a program to simulate the rolling of a fair six-sided die and calculate the expected value and variance of the outcomes.
import random
import statistics

class DiceSimulator:
    def __init__(self):
        # Outcomes of a fair six-sided die
        self.outcomes = [1, 2, 3, 4, 5, 6]
        # Probabilities for a fair die (each outcome has an equal probability of 1/6)
        self.probabilities = [1/6] * 6
    
    def roll_die(self, num_rolls):
        """Simulates rolling a six-sided die for num_rolls times."""
        rolls = [random.choice(self.outcomes) for _ in range(num_rolls)]
        return rolls

    def calculate_expected_value(self):
        """Calculates the expected value of the outcomes of a fair six-sided die."""
        expected_value = sum(outcome * prob for outcome, prob in zip(self.outcomes, self.probabilities))
        return expected_value

    def calculate_variance(self):
        """Calculates the variance of the outcomes of a fair six-sided die."""
        expected_value = self.calculate_expected_value()
        variance = sum(prob * ((outcome - expected_value) ** 2) for outcome, prob in zip(self.outcomes, self.probabilities))
        return variance


# Create an instance of the DiceSimulator
dice_simulator = DiceSimulator()

# Simulate rolling the die 1000 times
num_rolls = 1000
rolls = dice_simulator.roll_die(num_rolls)

# Calculate the expected value and variance of the outcomes
expected_value = dice_simulator.calculate_expected_value()
variance = dice_simulator.calculate_variance()

# Output the results
print(f"Simulated {num_rolls} rolls of a fair six-sided die.")
print(f"Expected Value: {expected_value}")
print(f"Variance: {variance}")

# Calculate the sample variance from the simulated rolls
sample_variance = statistics.variance(rolls)
print(f"Sample Variance from {num_rolls} rolls: {sample_variance}")
````

````python
#Question5:_) Create a Python function to generate random samples from a given probability distribution (e.g., binomial, Poisson) and calculate their mean and variance? 
import numpy as np
import scipy.stats as stats

# Function to generate random samples and calculate mean and variance
def generate_random_samples(distribution, params, sample_size=1000):
    """
    Generates random samples from a given distribution and calculates mean and variance.
    
    Args:
    - distribution (str): The type of distribution ('binomial', 'poisson').
    - params (tuple): Parameters for the distribution.
        For binomial: (n, p)
        For poisson: (lambda,)
    - sample_size (int): The number of samples to generate (default=1000).
    
    Returns:
    - mean (float): The mean of the generated samples.
    - variance (float): The variance of the generated samples.
    """
    if distribution == 'binomial':
        # Unpack the parameters (n, p)
        n, p = params
        # Generate random samples from the binomial distribution
        samples = np.random.binomial(n=n, p=p, size=sample_size)
    
    elif distribution == 'poisson':
        # Unpack the parameter (lambda)
        lam = params[0]
        # Generate random samples from the Poisson distribution
        samples = np.random.poisson(lam=lam, size=sample_size)
    
    else:
        raise ValueError("Unsupported distribution type. Use 'binomial' or 'poisson'.")
    
    # Calculate the mean and variance of the samples
    mean = np.mean(samples)
    variance = np.var(samples)
    
    return mean, variance

# Example Usage:

# 1. Binomial Distribution (n=10, p=0.5)
binomial_mean, binomial_variance = generate_random_samples('binomial', (10, 0.5), sample_size=1000)
print(f"Binomial Distribution (n=10, p=0.5): Mean = {binomial_mean}, Variance = {binomial_variance}")

# 2. Poisson Distribution (lambda=3)
poisson_mean, poisson_variance = generate_random_samples('poisson', (3,), sample_size=1000)
print(f"Poisson Distribution (lambda=3): Mean = {poisson_mean}, Variance = {poisson_variance}")
````

````python
#Question 6: Write a Python script to generate random numbers from a Gaussian (normal) distribution and compute the mean, variance, and standard deviation of the samples
import numpy as np

# Function to generate random samples from a Gaussian distribution and compute statistics
def gaussian_statistics(mean, std_dev, num_samples):
    # Generate random samples from a Gaussian (normal) distribution
    samples = np.random.normal(loc=mean, scale=std_dev, size=num_samples)
    
    # Calculate the mean of the samples
    calculated_mean = np.mean(samples)
    
    # Calculate the variance of the samples
    calculated_variance = np.var(samples)
    
    # Calculate the standard deviation of the samples
    calculated_std_dev = np.std(samples)
    
    return samples, calculated_mean, calculated_variance, calculated_std_dev

# Example usage
mean = 0        # Mean of the Gaussian distribution
std_dev = 1     # Standard deviation of the Gaussian distribution
num_samples = 1000  # Number of random samples to generate

# Generate the random samples and compute statistics
samples, calculated_mean, calculated_variance, calculated_std_dev = gaussian_statistics(mean, std_dev, num_samples)

# Print the results
print(f"Generated Mean: {calculated_mean}")
print(f"Generated Variance: {calculated_variance}")
print(f"Generated Standard Deviation: {calculated_std_dev}")
````

````python
#Question 11:
'''A company wants to test if a new website layout leads to a higher conversion rate (percentage of visitors 
who make a purchase). They collect data from the old and new layouts to compare.  
To generate the data use the following command: 
```python 
import numpy as np 
# 50 purchases out of 1000 visitors 
old_layout = np.array([1] * 50 + [0] * 950) 
# 70 purchases out of 1000 visitors   
new_layout = np.array([1] * 70 + [0] * 930) 
  ``` 
Apply z-test to find which layout is successful.'''
````

````python
"""State the Hypotheses:

Null Hypothesis 0: There is no difference between the old and new layout conversion rates. 
=
 
Alternative Hypothesis 1: The new layout has a higher conversion rate than the old layout. 
>(one-tailed test).
"""

import numpy as np
from statsmodels.stats.proportion import proportions_ztest

# Generate data for old and new layouts
old_layout = np.array([1] * 50 + [0] * 950)  # 50 purchases, 1000 visitors
new_layout = np.array([1] * 70 + [0] * 930)  # 70 purchases, 1000 visitors

# Number of purchases
x_old = sum(old_layout)
x_new = sum(new_layout)

# Total visitors
n_old = len(old_layout)
n_new = len(new_layout)

# Z-test for two proportions
count = np.array([x_old, x_new])
nobs = np.array([n_old, n_new])
z_stat, p_value = proportions_ztest(count, nobs, alternative='larger')

print(f"Z-statistic: {z_stat}")
print(f"P-value: {p_value}")

# Decision based on p-value
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The new layout leads to a higher conversion rate.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in conversion rates.")
````

````python
#Question 12:
''' A tutoring service claims that its program improves students' exam scores. A sample of students who 
participated in the program was taken, and their scores before and after the program were recorded.  
Use the below code to generate samples of respective arrays of marks: 
```python 
before_program = np.array([75, 80, 85, 70, 90, 78, 92, 88, 82, 87]) 
after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88]) 
``` 
Use z-test to find if the claims made by tutor are true or false.'''

#Anse:
"""To determine if the tutoring program significantly improved the students' exam scores, 
I can perform a paired z-test (also known as a two-sample z-test) on the scores before and 
after the program. The paired z-test is appropriate when you have two related samples 
(e.g., the same students' scores before and after the intervention)."""

import numpy as np
import scipy.stats as stats

# Sample data
before_program = np.array([75, 80, 85, 70, 90, 78, 92, 88, 82, 87])
after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88])

# Calculate differences
differences = after_program - before_program

# Mean and standard deviation of the differences
mean_diff = np.mean(differences)
std_diff = np.std(differences, ddof=0)  # Population standard deviation
n = len(differences)  # Sample size

# Calculate the Z-score
z_score = mean_diff / (std_diff / np.sqrt(n))

# Calculate the p-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Significance level
alpha = 0.05

# Print results
print(f"Z-score: {z_score}")
print(f"P-value: {p_value}")

if p_value < alpha:
    print("The tutoring program significantly improved students' scores.")
else:
    print("There is no significant improvement in students' scores.")
````

````python
#Question 13:
''' A pharmaceutical company wants to determine if a new drug is effective in reducing blood pressure. They 
conduct a study and record blood pressure measurements before and after administering the drug.  
Use the below code to generate samples of respective arrays of blood pressure:  
```python 
before_drug = np.array([145, 150, 140, 135, 155, 160, 152, 148, 130, 138]) 
after_drug = np.array([130, 140, 132, 128, 145, 148, 138, 136, 125, 130]) 
  ```  
Implement z-test to find if the drug really works or not. 
'''

import numpy as np
import scipy.stats as stats

# Blood pressure data
before_drug = np.array([145, 150, 140, 135, 155, 160, 152, 148, 130, 138])
after_drug = np.array([130, 140, 132, 128, 145, 148, 138, 136, 125, 130])

# Step 1: Calculate the differences
differences = before_drug - after_drug

# Step 2: Calculate the mean and standard deviation of the differences
mean_diff = np.mean(differences)
std_diff = np.std(differences, ddof=0)  # Population standard deviation
n = len(differences)  # Sample size

# Step 3: Calculate the Z-score
z_score = mean_diff / (std_diff / np.sqrt(n))

# Step 4: Calculate the P-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Step 5: Print results
alpha = 0.05  # Significance level

print(f"Z-score: {z_score}")
print(f"P-value: {p_value}")

if p_value < alpha:
    print("The drug significantly reduces blood pressure.")
else:
    print("The drug does not significantly reduce blood pressure.")
````

````python
#Question 14:
''' A customer service department claims that their average response time is less than 5 minutes. A sample 
of recent customer interactions was taken, and the response times were recorded.  
Implement the below code to generate the array of response time: 
```python 
response_times = np.array([4.3, 3.8, 5.1, 4.9, 4.7, 4.2, 5.2, 4.5, 4.6, 4.4]) 
``` 
Implement z-test to find the claims made by customer service department are tru or false.'''

#Answer:
"""In this case, the customer service department claims that their average response time is less than 5 minutes. 
We can perform a one-sample z-test to determine whether this claim is statistically supported.

Null Hypothesis (H): The mean response time is greater than or equal to 5 minutes (5).
Alternative Hypothesis (H): The mean response time is less than 5 minutes (<5)."""

import numpy as np
import scipy.stats as stats

# Given data: Response times in minutes
response_times = np.array([4.3, 3.8, 5.1, 4.9, 4.7, 4.2, 5.2, 4.5, 4.6, 4.4])

# Step 1: Calculate the sample mean and standard deviation
sample_mean = np.mean(response_times)
sample_std = np.std(response_times, ddof=1)  # Sample standard deviation
n = len(response_times)  # Sample size

# Step 2: Set the population mean (claimed mean)
population_mean = 5  # Claimed response time in minutes

# Step 3: Calculate the Z-score
z_score = (sample_mean - population_mean) / (sample_std / np.sqrt(n))

# Step 4: Calculate the P-value (one-tailed test for mean < population mean)
p_value = stats.norm.cdf(z_score)

# Step 5: Set significance level
alpha = 0.05  # 5% significance level

# Print results
print(f"Sample Mean: {sample_mean}")
print(f"Z-score: {z_score}")
print(f"P-value: {p_value}")

if p_value < alpha:
    print("Reject the null hypothesis. The customer service department's claim is true (average response time is less than 5 minutes).")
else:
    print("Fail to reject the null hypothesis. The customer service department's claim is not supported (average response time is not significantly less than 5 minutes).")
````

````python
#Question 15:
''' A company is testing two different website layouts to see which one leads to higher click-through rates. 
Write a Python function to perform an A/B test analysis, including calculating the t-statistic, degrees of 
freedom, and p-value.  
Use the following data: 
```python 
layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37] 
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]'''


import numpy as np
import scipy.stats as stats

# Given data: Click-through rates for two layouts
layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

# Step 1: Calculate the sample means and standard deviations
mean_a = np.mean(layout_a_clicks)
mean_b = np.mean(layout_b_clicks)
std_a = np.std(layout_a_clicks, ddof=1)
std_b = np.std(layout_b_clicks, ddof=1)

# Step 2: Calculate the sample sizes
n_a = len(layout_a_clicks)
n_b = len(layout_b_clicks)

# Step 3: Calculate the T-statistic and P-value using an independent two-sample t-test
t_stat, p_value = stats.ttest_ind(layout_a_clicks, layout_b_clicks)

# Step 4: Calculate the degrees of freedom
df = (n_a - 1) + (n_b - 1)

# Print results
print(f"Mean of Layout A Clicks: {mean_a}")
print(f"Mean of Layout B Clicks: {mean_b}")
print(f"T-Statistic: {t_stat}")
print(f"Degrees of Freedom: {df}")
print(f"P-Value: {p_value}")

# Step 5: Conclusion based on p-value
alpha = 0.05  # 5% significance level
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in click-through rates between Layout A and Layout B.")
else:
    print("Fail to reject the null hypothesis. No significant difference in click-through rates between Layout A and Layout B.")
````

````python
#Question 17:
'''A school district introduces an educational intervention program to improve math scores. Write a Python 
function to analyze pre- and post-intervention test scores, calculating the t-statistic and p-value to 
determine if the intervention had a significant impact.  
Use the following data of test score:  
  ```python 
  pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87] 
  post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93] 
  ``` 
'''

import numpy as np
from scipy import stats

def analyze_intervention(pre_scores, post_scores):
    # Step 1: Calculate the mean of both groups
    mean_pre = np.mean(pre_scores)
    mean_post = np.mean(post_scores)
    
    print(f"Mean Pre-Intervention Scores: {mean_pre}")
    print(f"Mean Post-Intervention Scores: {mean_post}")
    
    # Step 2: Perform a paired t-test
    t_statistic, p_value = stats.ttest_rel(pre_scores, post_scores)
    
    # Step 3: Output the t-statistic and p-value
    print(f"T-statistic: {t_statistic}")
    print(f"P-value: {p_value}")
    
    # Step 4: Interpret the result
    alpha = 0.05  # Significance level
    if p_value < alpha:
        print("Reject the null hypothesis. The intervention had a significant impact on math scores.")
    else:
        print("Fail to reject the null hypothesis. The intervention did not have a significant impact on math scores.")

# Data: Pre- and post-intervention test scores
pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87]
post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]

# Perform the analysis
analyze_intervention(pre_intervention_scores, post_intervention_scores)
````

````python
#Question 21:
''' A political analyst wants to determine if there is a significant association between age groups and voter 
preferences  (Candidate A or Candidate B). They collect data from a sample of 500 voters and classify 
them into different age groups and candidate preferences. Perform a Chi-S quare test to determine if 
there is a significant association between age groups and voter preferences.  
Use the below code to generate data: 
```python 
np.random.seed(0) 
age_groups = np.random.choice(['18 -30', '31 -50', '51 +', '51 +'], size=30) 
voter_preferences = np.random.choice(['C andidate A', 'C andidate B'] , size=30) 
'''

#Answer:
"""The Chi-Square test is a statistical test used to determine if there is a significant 
association between categorical variables. It assesses whether the observed frequencies 
in a contingency table differ from the frequencies we would expect if there were no association between the variables"""

import numpy as np
import pandas as pd
from scipy import stats

# Generate synthetic data
np.random.seed(0)
age_groups = np.random.choice(['18-30', '31-50', '51+'], size=500)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=500)

# Create a DataFrame
data = pd.DataFrame({'Age Group': age_groups, 'Voter Preference': voter_preferences})

# Create a contingency table
contingency_table = pd.crosstab(data['Age Group'], data['Voter Preference'])

# Print the contingency table
print("Contingency Table:")
print(contingency_table)

# Perform Chi-Square test
chi2_stat, p_value, dof, expected = stats.chi2_contingency(contingency_table)

# Print the results
print(f"\nChi-Square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:")
print(expected)

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("\nThere is a significant association between age groups and voter preferences.")
else:
    print("\nThere is no significant association between age groups and voter preferences.")
````

````python
#Question 24:
''' A company produces three different versions of a product:  Standard, Premium, and Deluxe. The 
company wants to determine if there is a significant difference in customer satisfaction scores among the 
three product versions. They conducted a survey and collected customer satisfaction scores for each 
version from a random sample of customers. Perform an ANOV A test to determine if there is a significant 
difference in customer satisfaction scores.  
  Use the following data: 
  ```python 
  # Sample data: Customer satisfaction scores for each product version 
  standard_ scores = [80, 85, 90, 78, 88, 82, 92, 78, 85, 87] 
  premium _ scores = [90, 92, 88, 92, 95, 91 , 96, 93, 89, 93] 
  deluxe _ scores = [95, 98, 92, 97, 96, 94, 98, 97, 92, 99] 
  ``` '''

#Answer:
"""
ANOVA (Analysis of Variance) is a statistical method used to test whether there 
are significant differences between the means of three or more independent groups. 
It helps determine if the observed differences in sample means are likely due to 
real differences between the populations from which the samples were drawn or if they are simply due to random variation.

<<Steps to Perform ANOVA>>

1.Formulate Hypotheses:
Null Hypothesis (H): All group means are equal.
Alternative Hypothesis (H): At least one group mean is different.

2.Calculate the ANOVA Test Statistic:
Compute the between-group and within-group variances.
Calculate the F-statistic using the ratio of these variances.

3.Determine the p-value:
Compare the F-statistic with the F-distribution to determine the p-value.

4.Make a Decision:
If the p-value is less than the significance level (e.g., 0.05), reject the null hypothesis, indicating that there are significant differences between group means."""

import numpy as np
from scipy.stats import f_oneway

# Sample data: Customer satisfaction scores for each product version
standard_scores = [80, 85, 90, 78, 88, 82, 92, 78, 85, 87]
premium_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]
deluxe_scores = [95, 98, 92, 97, 96, 94, 98, 97, 92, 99]

# Perform ANOVA test
f_statistic, p_value = f_oneway(standard_scores, premium_scores, deluxe_scores)

# Output the results
print(f"F-Statistic: {f_statistic}")
print(f"P-Value: {p_value}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("There is a significant difference in customer satisfaction scores among the product versions.")
else:
    print("There is no significant difference in customer satisfaction scores among the product versions.")
````
